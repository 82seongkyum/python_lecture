##### 스타일 지정하기

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo-')
plt.axis([1, 6, 0, 20])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 200ms 간격으로 균일하게 샘플된 시간

t = np.arange(0, 10, 0.2)
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'b-')
plt.plot(t, t**3, 'g-.')
plt.show()
```
